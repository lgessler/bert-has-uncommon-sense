```python
from ldg.pickle import pickle_read
```

```python
data5 = pickle_read('predictions/bert-base-cased.tsv.5.pkl')
data10 = pickle_read('predictions/bert-base-cased.tsv.10.pkl')
data15 = pickle_read('predictions/bert-base-cased.tsv.15.pkl')
data20 = pickle_read('predictions/bert-base-cased.tsv.20.pkl')
data25 = pickle_read('predictions/bert-base-cased.tsv.25.pkl')
data30 = pickle_read('predictions/bert-base-cased.tsv.30.pkl')
data35 = pickle_read('predictions/bert-base-cased.tsv.35.pkl')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
plt.plot([x['synset'] for x in data5.values()], color='red', marker='o')
plt.plot([x['synset'] for x in data10.values()], color='orange', marker='o')
plt.plot([x['synset'] for x in data15.values()], color='yellow', marker='o')
plt.plot([x['synset'] for x in data20.values()], color='green', marker='o')
plt.plot([x['synset'] for x in data25.values()], color='blue', marker='o')
plt.plot([x['synset'] for x in data30.values()], color='indigo', marker='o')
plt.plot([x['synset'] for x in data35.values()], color='violet', marker='o')

plt.plot([x['label'] for x in data5.values()], color='red', marker='x')
plt.plot([x['label'] for x in data10.values()], color='orange', marker='x')
plt.plot([x['label'] for x in data15.values()], color='yellow', marker='x')
plt.plot([x['label'] for x in data20.values()], color='green', marker='x')
plt.plot([x['label'] for x in data25.values()], color='blue', marker='x')
plt.plot([x['label'] for x in data30.values()], color='indigo', marker='x')
plt.plot([x['label'] for x in data35.values()], color='violet', marker='x')
plt.ylabel('precision')
plt.xlabel('k-1')
plt.ylim(0.3,1)
plt.xlim(-1,50)
plt.show()
```
